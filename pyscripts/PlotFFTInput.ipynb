```python
import numpy as np
import matplotlib.pyplot as plt
import csv
import CommonScript as cs
plt.rcParams['figure.figsize']=[10,10]
```

```python
with open('./TransiverFFTInput.csv','r') as fp:
    csvBuffer = list(csv.reader(fp))
    realInput =  np.array(csvBuffer[0],dtype=int)[::3]/512
    imagInput = np.array(csvBuffer[1],dtype=int)[::3]/512
```

```python
demodOut = cs.ConstellationDeMapping(realInput,imagInput)

print(hex(int(demodOut,2)))
```

```python
complexInput = realInput+imagInput*1j
plt.subplot(211)
plt.plot(np.linspace(1,256,256),np.real(np.fft.ifft(complexInput)))
plt.title('theory ifft of 0x12345678...pattern:real ')
plt.subplot(212)
plt.title('theory ifft of 0x12345678...pattern:imag ')

plt.plot(np.linspace(1,256,256),np.imag(np.fft.ifft(complexInput)))
```
