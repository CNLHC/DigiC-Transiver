```python
import numpy as np
import matplotlib.pyplot as plt
import csv
import functools
from bitstring import BitArray
plt.rcParams['figure.figsize']=[10,10]
```

```python
with open('./FFTDataApapterIn.csv','r') as fp:
    csvBuffer=list(csv.reader(fp))
    data = csvBuffer[0][::3]
```

```python
def QAMDeDemodulation(real,imag):
    bitstream=[]
    for r,i in zip(real,imag):
        if(r>0 and i>0):
            bitstream.append('00')
        elif (r<0 and i>0):
            bitstream.append('01')
        elif (r<0 and i<0):
            bitstream.append('11')
        elif (r>0 and i<0):
            bitstream.append('10')
    return ''.join(bitstream)
```

```python
bitstream = list(map(lambda x:"{0:s}".format(np.binary_repr(int(x),width=4)) ,data))
real = [BitArray(bin=i[0:2]).int for i in bitstream]
imag = [BitArray(bin=i[2:4]).int for i in bitstream]
demoudOut = QAMDeDemodulation(real,imag)
print(real)
print(imag)
print(bitstream)
print (demoudOut)
print(hex(int(demoudOut,2)))
```
